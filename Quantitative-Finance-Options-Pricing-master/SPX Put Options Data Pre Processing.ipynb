```python
from yahoo_fin import options
import datetime
from datetime import datetime,date
from wallstreet import Call, Put
import pandas as pd
```

```python
#!pip install wallstreet
# Get expirations Date 

SPX_Put = Put('^SPX')
expiration_date_list = SPX_Put.expirations
expiration_date_list
```

```python
chain = options.get_options_chain("^SPX")
chain['puts']
```

```python
SPX = pd.DataFrame()

#SPX = options.get_calls('^SPX', "13-12-2019")
#T0 = datetime.datetime.now().date()
#SPX['Exp_Date'] = date
#SPX
```

```python
def get_put_data(SPX, tickr, exp_dates):
    #SPX = pd.DataFrame()
    T0 = datetime.date(datetime.now())
    for date in exp_dates:
        print('Fetching Options Data for Date - '+ date)
        #SPX_temp = pd.DataFrame()
        try:

            SPX_temp = options.get_puts(tickr, date)
            #print(SPX_temp)
            SPX_temp['Exp_Date'] = date
            #SPX.append(SPX_temp, ignore_index= False)
            SPX = pd.concat([SPX, SPX_temp])
        except:
            print('No data found for ' + date)
    
    SPX['Current Date'] = T0
    return SPX
```

```python
def date_format(date_time_str):

    #date_time_str = '2019-12-20'
    date_time_obj = datetime.datetime.strptime(date_time_str, '%Y-%m-%d')

    print('Date:', date_time_obj.date())
    print('Time:', date_time_obj.time())
    print('Date-time:', date_time_obj)
    formdate = date_time_obj.date()
    return formdate
```

```python
#SPX = pd.DataFrame()
SPX_Puts = get_put_data(SPX, '^SPX', expiration_date_list)
```
